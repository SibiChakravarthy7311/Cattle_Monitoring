```python
# IMPORT STATEMENTS
import os
import random
import shutil
from ultralytics import YOLO
import numpy as np
import matplotlib.pyplot as plt
import torch
from pathlib import Path
```

```python
# FUNCTION TO TRAIN AND TEST MODELS
def trainAndTestModel(modelPath, dataYaml, saveDir, epochs=100, imgSize=640, testSource="/content/drive/MyDrive/A36_Training_Files/A36_Cows_Extracted_YOLO/images/val"):
    model = YOLO(modelPath)
    print(f"Starting training for model: {modelPath}")
    model.train(data=dataYaml, epochs=epochs, imgsz=imgSize)

    trained_model_path = f"{saveDir}/{modelPath.split('/')[-1].replace('.pt', '')}_400.pt"
    model.save(trained_model_path)
    print(f"Model saved to {trained_model_path}")

    print(f"Starting inference for model: {modelPath}")
    result = model.predict(source=testSource)

    print(f"Inference completed for model: {modelPath}")
    return result
```

```python
# INITIALIZING THE LIST OF MODELS TO BE TRAINED IN THIS NOTEBOOK
modelList = ["/content/drive/MyDrive/A36_Training_Files/A36_Cows_Extracted_YOLO/Trained_Models/yolo11x-pose_trained_trained_200.pt"]
```

```python

# TRAIN AND TEST MODELS IN THE MODEL LIST
dataYaml = "/content/drive/MyDrive/A36_Training_Files/A36_Cows_Extracted_YOLO/cow_data.yaml"
epochs = 200
saveDir = "/content/drive/MyDrive/A36_Training_Files/A36_Cows_Extracted_YOLO/Trained_Models/"
os.makedirs(saveDir, exist_ok=True)
for modelPath in modelList:
    print("----------------------------------------------------------")
    print(f"Processing Model: {modelPath}")
    trainAndTestModel(modelPath, dataYaml, saveDir, epochs=epochs)
    print("----------------------------------------------------------")
```
